```python
# Function to generate summary stats for numeric columns in dataset/dataframe.
def SummaryStats(column, cleaned_df):
    mean_numpy = np.mean(cleaned_df[column])
    median_numpy = np.median(cleaned_df[column])
    min_numpy = np.min(cleaned_df[column])
    max_numpy = np.max(cleaned_df[column])
    std_numpy = np.std(cleaned_df[column])
    var_numpy = np.var(cleaned_df[column])
    sem_numpy = cleaned_df[column].sem()

    summary_df = pd.DataFrame({ f"Average Number of {column}":round(mean_numpy,2),
                                f"Median Number of {column}":round(median_numpy,2),
                                f"Minimum of {column}":round(min_numpy,2),
                                f"Maximum of {column}":round(max_numpy,2),
                                f"Standard Deviation of {column}":round(mean_numpy,2),
                                f"Variance of {column}":round(median_numpy,2),
                                f"Standard Error of Mean of {column}":round(sem_numpy,2)},index = [0])

    return summary_df

# Function to generate histogram
def Histogram(column, cleaned_df, number_bins, title):
    plt.figure(figsize=(12, 5))
    plt.hist(cleaned_df[column], bins=number_bins)
    plt.xlabel(f'Star {column}')
    plt.ylabel('Number of Restaurants')
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
below_2k = []
from_2k_to_4k = []
above_4k = []

counter = 0
for i in cleaned_data["Review Count"]:
    if i <= 2000:
        below_2k.append(i)
    if i > 2000 and i <= 4000:
        from_2k_to_4k.append(i)
    if i > 4000:
        above_4k.append(i)
        counter += 1

print(f"There are {len(below_2k)} restaurants with less than or equal to 2000 reviews\
, {len(from_2k_to_4k)} restaurants with 2000-4000 reviews\
, {len(above_4k)} restaurants with greater than 4000 reviews.")


below_2k_sample = cleaned_data.loc[(cleaned_data["Review Count"] <= 2000)].reset_index(drop=True)
Histogram("Review Count", below_2k_sample, 200, title="Restaurants Review Count")

plt.show()
```

```python
below_50 = []
from_50_to_200 = []
from_201_to_2000 = []

counter = 0
for i in below_2k_sample["Review Count"]:
    if i <= 50:
        below_50.append(i)
    if i > 50 and i <= 200:
        from_50_to_200.append(i)
    if i > 200:
        from_201_to_2000.append(i)
        counter += 1

print(f"There are {len(below_50)} restaurants with less than or equal to 50 reviews\
, {len(from_50_to_200)} restaurants with 50-200 reviews\
, {len(from_201_to_2000)} restaurants with greater than 200 and less than 2000 reviews\
, {len(from_2k_to_4k)+len(above_4k)} restaurants with greater than 2000 reviews.")
```

```python
# Prepare the ANOVA data, assume 5 levels, based on the number of restaurants in each review count category, loaded in the factor "Review Count"
all_levels = below_50 + from_50_to_200 + from_201_to_2000 + from_2k_to_4k + above_4k
data_levels = (["below_50"]*len(below_50) + ["from_50_to_200"]*len(from_50_to_200)
       + ["from_201_to_2000"]*len(from_201_to_2000) + ["from_2k_to_4k"]*len(from_2k_to_4k)
       + ["above_4k"]*len(above_4k))

df_levels = pd.DataFrame({"Review Count": all_levels, "Levels": data_levels})

df_levels.head()
```

```python
df_levels.groupby("Levels").mean()
```

```python
# Hypothesis testing
# Null hypothesis: the means of each levels are not significantly different
# Alternative hypothesis: the means of each levels are significantly different

# one-way ANOVA
st.f_oneway(df_levels['Review Count'][df_levels['Levels'] == "below_50"],
               df_levels['Review Count'][df_levels['Levels'] == "from_50_to_200"],
               df_levels['Review Count'][df_levels['Levels'] == "from_201_to_2000"],
               df_levels['Review Count'][df_levels['Levels'] == "from_2k_to_4k"],
               df_levels['Review Count'][df_levels['Levels'] == "above_4k"])

# Choose alpha level = 0.05
# It is noted that  p-value is below 0.05 (and that F is very large), we reject the null hypothesis in favor of the alternative hypotheis 
# this means that at least one group mean is significantly different.
```

```python
# Summary output of ANOVA statistical analysis (using researchpy module)
rp.summary_cont(df_levels['Review Count'].groupby(df_levels['Levels']))
```
